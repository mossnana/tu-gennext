```
def is_leap_year(year):
  if year % 400 == 0:
    print("Leap Year !!!")
  elif year % 4 == 0 and year % 100 == 0:
    print("Leap Year !!!")
  else:
    print("Not Leap Year !!!")

def main():
  is_leap_year(1900)
  is_leap_year(2011)
  is_leap_year(2100)
  is_leap_year(2000)
  is_leap_year(2008)
  is_leap_year(2018)
  
main()
```

```
import random
def get_random():
  message1 = "Amazing"
  message2 = "Romance"
  message3 = "Proceed"
  message4 = "Lucky"
  randomInt = random.randrange(1, 11)
  if randomInt < 5:
    print(message1)
    return 1
  elif randomInt < 8:
    print(message2)
    return 2
  elif randomInt < 10:
    print(message3)
    return 3
  else:
    print(message4)
    return 4

def main():
  i = 0
  status = 0
  store = [0, 0, 0, 0]
  while i < 11:
    status = get_random()
    if status == 1:
      store[0] += 1
    elif status == 2:
      store[1] += 1
    elif status == 3:
      store[2] += 1
    elif status == 4:
      store[3] += 1
    i += 1
  print(store)
  print("Finish Random")
  
main()
```
